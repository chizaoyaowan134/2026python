# ndarry与python列表运算效率比较

```python
import random
import time

a = []

for i in range(100000000):
    a.append(random.random())

print("随机列表生成完成")

t1 = time.time()
sum1 = sum(a)
t2 = time.time()

b = np.array(a)
print("随机数组生成完成")
t4 = time.time()
sum2 = np.sum(b)
t5 = time.time()

print("列表求和用时:", t2 - t1)
print("数组求和用时:", t5 - t4)
```
