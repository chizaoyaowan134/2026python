请你总结上述广播机制的规则
1. 如果两个数组的维度数不同，那么维度较少的数组会在前面补1，直到两个数组的维度数相同。
2. 如果两个数组在某个维度上的长度不同，但其中一个数组在该维度上的长度为1，那么该数组会沿着该维度进行扩展，以匹配另一个数组的长度。
3. 如果两个数组在某个维度上的长度都不相同且都不为1，那么会引发错误，表示无法进行广播。
4. 广播机制允许在不复制数据的情况下进行数组运算，从而提高计算效率。
5. 广播机制适用于各种算术运算，如加法、减法、乘法和除法等。
6. 广播机制在处理多维数组时尤为有用，能够简化代码并提高可读性。
7. 理解广播机制有助于更好地利用NumPy进行高效的数值计算。
8. 广播机制不仅适用于数组与数组之间的运算，也适用于数组与标量之间的运算。
9. 广播机制在数据分析和科学计算中广泛应用，是NumPy的核心特性之一。
10. 掌握广播机制有助于优化代码性能，减少内存使用，提高计算速度。
11. 广播机制在处理不同形状的数组时提供了灵活性，使得代码更加简洁和易于维护。

```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print("-" * 20)
print(np.sum(a, axis=0)) # 列求和,结果是每一列的和,返回一个一维数组,长度等于列数
print("-" * 20)
print(np.sum(a, axis=1)) # 行求和,结果是每一行的和,返回一个一维数组,长度等于行数
```

```python
# import numpy as np

# 创建一个形状为 (2, 2, 2, 2) 的四维数组
# 可以理解为：2个大组，每个大组里有2张表，每张表是2x2
a = np.arange(16).reshape(2, 2, 2, 2)

# 执行 axis=0 降维求和
b = np.sum(a, axis=0)

print(a)
print("-" * 20)
print("Shape after summing over axis 0:", b.shape)  # 结果形状应为 (2, 2, 2)
print(b)
print("-" * 20)
# 执行 axis=1 降维求和
c = np.sum(a, axis=1)
print("Shape after summing over axis 1:", c.shape)  # 结果形状应为 (2, 2, 2)
print(c)
print("-" * 20)
# 执行 axis=2 降维求和
d = np.sum(a, axis=2)
print("Shape after summing over axis 2:", d.shape)  # 结果形
print(d)
print("-" * 20)
# 执行 axis=3 降维求和
e = np.sum(a, axis=3)
print("Shape after summing over axis 3:", e.shape)  # 结果形
print(e)
```
